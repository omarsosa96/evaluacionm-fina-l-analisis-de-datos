b. Seleccionar una base de entrenamiento (80%) y una base de prueba (20%).

```python
# prompt: . Seleccionar una base de entrenamiento (80%) y una base de prueba (20%).

# Definir la proporción de la muestra de entrenamiento
train_ratio = 0.7

# Calcular el tamaño de la muestra de entrenamiento
train_size = int(len(Datos) * train_ratio)

# Dividir los datos en conjuntos de entrenamiento y prueba
train_data = Datos.iloc[:train_size]
test_data = Datos.iloc[train_size:]

# Imprimir los tamaños de los conjuntos de entrenamiento y prueba
print(f"Tamaño del conjunto de entrenamiento: {len(train_data)}")
print(f"Tamaño del conjunto de prueba: {len(test_data)}")
```

c. Con la base de entrenamiento aplicar un modelo de regresión lineal para predecir Calificación promedio (Aggregate rating) a partir de: costo (Average Cost for two),
número de calificaciones (Votes) y Pedido online (Has Online delivery).

```python
# Convertir la columna 'Has Online delivery' a numérica (0 y 1)
train_data['Has Online delivery'] = train_data['Has Online delivery'].map({'Yes': 1, 'No': 0})

# Seleccionar las características (X) y la variable objetivo (y)
X = train_data[['Average Cost for two', 'Votes', 'Has Online delivery']]
y = train_data['Aggregate rating']

# Dividir los datos de entrenamiento en conjuntos de entrenamiento y validación
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.3, random_state=42)

# Crear y entrenar el modelo de regresión lineal
model = LinearRegression()
model.fit(X_train, y_train)

# Hacer predicciones en el conjunto de validación
y_pred = model.predict(X_val)

# Evaluar el modelo
mse = mean_squared_error(y_val, y_pred)
r2 = r2_score(y_val, y_pred)

print(f"Error cuadrático medio (MSE): {mse}")
print(f"Coeficiente de determinación (R^2): {r2}")
```

d. Realizar las predicciones sobre la base de prueba. Calcule el R², MSE y RMSE.

```python

# Seleccionar las características (X) y la variable objetivo (y) del conjunto de prueba
X_test = train_data[['Average Cost for two', 'Votes', 'Has Online delivery']]
y_test = train_data['Aggregate rating']

# Realizar predicciones en el conjunto de prueba
y_pred_test = model.predict(X_test)

# Calcular las métricas de evaluación en el conjunto de prueba
mse_test = mean_squared_error(y_test, y_pred_test)
r2_test = r2_score(y_test, y_pred_test)
rmse_test = np.sqrt(mse_test)

print(f"Error cuadrático medio (MSE) - Test: {mse_test}")
print(f"Coeficiente de determinación (R^2) - Test: {r2_test}")
print(f"Raíz del error cuadrático medio (RMSE) - Test: {rmse_test}")
```

b. Construir la variable ‘nivel_calificacion’ con niveles alto y bajo. Considere nivel alto
a calificaciones promedio > 3.7 y nivel bajo a calificaciones promedio <=3.7. En
caso de tener la variable objetivo con un desbalanceo, utilizar el comando
iloc[random_index] tanto en X como en y


```python
# Definir la función para crear la variable 'nivel_calificacion'
def asignar_nivel(calificacion):
    if calificacion > 3.7:
        return 'alto'
    else:
        return 'bajo'

# Aplicar la función a la columna 'Aggregate rating'
Datos['nivel_calificacion'] = Datos['Aggregate rating'].apply(asignar_nivel)

# Verificar el balanceo de la variable objetivo
nivel_counts = Datos['nivel_calificacion'].value_counts()
print(nivel_counts)

# Si hay desbalanceo, realizar el muestreo aleatorio
if nivel_counts['alto'] != nivel_counts['bajo']:
    # Obtener los índices de las clases mayoritarias y minoritarias
    indices_alto = Datos[Datos['nivel_calificacion'] == 'alto'].index
    indices_bajo = Datos[Datos['nivel_calificacion'] == 'bajo'].index

    # Obtener el tamaño de la clase minoritaria
    min_size = min(len(indices_alto), len(indices_bajo))

    # Muestreo aleatorio de la clase mayoritaria
    random_indices_alto = np.random.choice(indices_alto, size=min_size, replace=False)

    # Combinar los índices de la clase minoritaria y los índices aleatorios de la clase mayoritaria
    combined_indices = np.concatenate([random_indices_alto, indices_bajo])

    # Crear el nuevo DataFrame balanceado
    Datos_balanceados = Datos.iloc[combined_indices]

    # Imprimir el balanceo del nuevo DataFrame
    nivel_counts_balanceados = Datos_balanceados['nivel_calificacion'].value_counts()
    print(nivel_counts_balanceados)

    # Separar características (X) y variable objetivo (y) usando el DataFrame balanceado
    X = Datos_balanceados[['Average Cost for two', 'Votes', 'Has Online delivery']]  # Ajustar columnas según sea necesario
    y = Datos_balanceados['nivel_calificacion']
else:
    # Si no hay desbalanceo, usar los datos originales
    X = Datos[['Average Cost for two', 'Votes', 'Has Online delivery']] # Ajustar columnas según sea necesario
    y = Datos['nivel_calificacion']
```
